```python
import numpy as np
```

```python
x = np.array([
    [1, 0, 1, 1],
    [0, 0, 0, 1],
    [1, 0, 0, 1],
    [1, 1, 1, 0]
])
```

```python
w = np.array([
    [0.9, 0.3, 0.1, 0.2],
    [0.1, 0.6, 0.2, 0.1]
])
```

```python
lr = 0.1
```

```python
epochs = 1
for e in range(epochs):
    print(f"\nEpoch {e+1}")
    print(f"Learning Rate: ",lr)
    for i in range(len(x)):
        print(f"\nInput {i+1}: {x[i]}")

        distances = ((w - x[i])**2).sum(axis=1)

        bmu = np.argmin(distances)

        print("Distances:",distances)
        print("BMU:",bmu+1)

        w[bmu] = w[bmu] + lr*(x[i] - w[bmu])

        print("Updated Weights:\n",w)

    lr *= 0.5
```
