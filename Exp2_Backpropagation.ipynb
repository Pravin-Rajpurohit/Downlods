```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(O):
    return O * (1 - O)
```

```python
x1, x2 = 1, 0
w1, w2, w3, w4, w5, w6 = 0.5, 0.4, -0.3, 0.2, 0.1, 0.3
b1, b2, b3 = 0.3, -0.2, 0.5
d = 1
lr = 1

epoch = 0
mse = 1

while epoch < 500 and mse > 0:
    h1_in = x1 * w1 + x2 * w2 + b1
    h2_in = x1 * w3 + x2 * w4 + b2
    h1_out = sigmoid(h1_in)
    h2_out = sigmoid(h2_in)
    o_in = h1_out * w5 + h2_out * w6 + b3
    o_out = sigmoid(o_in)

    mse = 0.5 * (d - o_out) ** 2
    
    del_o = (d - o_out) * sigmoid_derivative(o_out)
    del_h1 = del_o * w5 * sigmoid_derivative(h1_out)
    del_h2 = del_o * w6 * sigmoid_derivative(h2_out)
    
    w5 = w5 + lr * del_o * h1_out
    w6 = w6 + lr * del_o * h2_out
    w1 = w1 + lr * del_h1 * x1
    w2 = w2 + lr * del_h1 * x2
    w3 = w3 + lr * del_h2 * x1
    w4 = w4 + lr * del_h2 * x2
    
    b1 = b1 + lr * del_h1
    b2 = b2 + lr * del_h2
    b3 = b3 + lr * del_o
    
    epoch += 1

    if epoch % 100 == 0 or epoch == 1:
        print(f"Epoch {epoch}: MSE = {mse:.10f}, Output = {o_out:.6f}")

print(f"Final MSE: {mse:.10f}")
print(f"Final Output: {o_out:.6f}")
print(f"\nFinal Weights:")
print(f"w1 = {w1:.6f}, w2 = {w2:.6f}, w3 = {w3:.6f}, w4 = {w4:.6f}, w5 = {w5:.6f}, w6 = {w6:.6f}")
print(f"\nFinal Biases:")
print(f"b1 = {b1:.6f}, b2 = {b2:.6f}, b3 = {b3:.6f}")
```

# XOR with Standard Gradient Descent

```python
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
Y = np.array([[0], [1], [1], [0]])

W1 = np.array([[0.5, -0.3], [0.4, 0.2]])
W2 = np.array([[0.1], [0.3]])
b1 = np.array([[0.3, -0.2]])
b2 = np.array([[0.5]])

lr = 1
for epoch in range(5000):
    H = sigmoid(np.dot(X, W1) + b1)
    O = sigmoid(np.dot(H, W2) + b2)
    
    del_O = (Y - O) * sigmoid_derivative(O)
    del_H = np.dot(del_O, W2.T) * sigmoid_derivative(H)
    
    W2 += lr * np.dot(H.T, del_O)
    b2 += lr * np.sum(del_O, axis=0, keepdims=True)
    W1 += lr * np.dot(X.T, del_H)
    b1 += lr * np.sum(del_H, axis=0, keepdims=True)

print("Outputs:", O.flatten())
```
