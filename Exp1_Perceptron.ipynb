```python
import numpy as np

def bipolar_activation(net):
    if net>0:
        act_val = 1
    elif net==0:
        act_val = 0
    else:
        act_val = -1
    return act_val

def unipolar_activation(net):
    if net>=0:
        act_val = 1
    else:
        act_val = 0
    return act_val

def train_perceptron(training_data, targets, w, b, lr, act_func):
    epoch = 0
    while True:
        epoch += 1
        del_w = False
        for i, (x, d) in enumerate(zip(training_data, targets)):
            net = np.dot(w, x) + b
            o = act_func(net)
            e = d - o
            if e != 0:
                del_w = True
                new_w = lr * e * x
                new_b = lr * e
                w = w + new_w
                b = b + new_b
        if not del_w:
            break
    return w, b
```

# AND

```python
print("Training Bipolar Perceptron")
bi_train = np.array([[-1, -1], [-1, 1], [1, -1], [1, 1]])
bi_targets = np.array([-1, -1, -1, 1])
bi_w = np.array([0.0, 0.0])
bi_b = 0.0
bi_lr = 1
final_bi_w, final_bi_b = train_perceptron(bi_train, bi_targets, bi_w, bi_b, bi_lr, bipolar_activation)
print(f"Final Bipolar Weights: {final_bi_w}, Final Bias: {final_bi_b}\n")

print("Training Unipolar Perceptron")
uni_train = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
uni_targets = np.array([0, 0, 0, 1])
uni_w = np.array([0.0, 0.0])
uni_b = 0.0
uni_lr = 1
final_uni_w, final_uni_b = train_perceptron(uni_train, uni_targets, uni_w, uni_b, uni_lr, unipolar_activation)
print(f"Final Unipolar Weights: {final_uni_w}, Final Bias: {final_uni_b}")
```

# Static Sum

```python
print("Static Sum")
uni_train = np.array([[0, 0], [1, 0], [0, 1], [1, 1]])

w_or, b_or = np.array([1.0, 1.0]), -0.5
w_nand, b_nand = np.array([-1.0, -1.0]), 1.5
w_and, b_and = np.array([1.0, 1.0]), -1.5

for x in uni_train:
    out_or = unipolar_activation(np.dot(w_or, x) + b_or)
    out_nand = unipolar_activation(np.dot(w_nand, x) + b_nand)
    hidden_out = np.array([out_or, out_nand])
    sum_out = unipolar_activation(np.dot(w_and, hidden_out) + b_and)
    print(f"Input: {x}, Sum: {sum_out}")
```
